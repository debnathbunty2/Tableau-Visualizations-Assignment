9. Q9. Chart Selection Justification

Explain why different charts were used for different business questions in this analysis.

- This dashboard uses three distinct chart types, each specifically chosen to optimize how the human brain processes different kinds of business data:
1. Cumulative Sales Growth:
Line Chart
Why it used:
Line charts are the standard for Time Series Analysis.
Business Insight: By using a running sum, the upward slope of the lines immediately shows the velocity of growth. It allows stakeholders to compare categories (represented by different colored lines) over time to see which is contributing most to the total volume and identify seasonal trends.
2. Product Prioritization:
Scatter Plot
Why it used:
Scatter plots are designed to show Correlations and Outliers between two numerical variables.
Business Insight:
By plotting Profit against Sales, the analyst can instantly identify "Star Products" (high sales, high profit) versus "Underperformers" (high sales, low profit). The addition of an "Average" reference line creates a quadrant for quick strategic decision-making.
3. Category vs Region Performance:
Highlight Table (Heat Map)
Why it used:
Highlight tables combine the precision of a Cross-tab with the visual cues of a Heat Map.
Business Insight:
While a standard table is good for looking up exact numbers (e.g., Furniture sales in the Central region), the color coding allows for pattern recognition. The eye is immediately drawn to the highest values (dark blue) and lowest values (dark orange), making regional performance gaps obvious without needing to read every single digit.

10. Q10. Business Insight Summary

Summarize two key insights from your dashboard and explain how they can help management make better
decisions

- 1. High Sales Growth with Regional Variations
The Cumulative Sales Growth chart shows steady upward momentum, ending January 2024 at over $329K. However, the Category vs Region Performance table reveals significant disparities: Technology is a powerhouse in the West ($114,990), while Furniture dominates the Central region ($100,032).
Management Decision: Use these insights to localize inventory and marketing. Instead of a "one-size-fits-all" national campaign, management should double down on Technology promotions in the West and Furniture-focused logistics in the Central region to maximize existing momentum.
2. Profit Concentration in High-Volume Products
The Product Prioritization scatter plot shows that most products cluster below the $40K profit mark. There is a clear outlier—a high-sales, high-profit product located in the top-right quadrant (near $240K sales).

Management Decision: This helps in Resource Allocation. Management can identify that specific "superstar" product for expansion or protection (e.g., ensuring supply chain stability). Conversely, for the products hovering around the "Average" sales line with lower profit, management may need to review pricing strategies or operational costs to improve their margins.
